# TRIBE v2 Persona-Based Reaction Predictor

This notebook demonstrates how to use the persona-based reaction prediction system built on Meta's TRIBE v2.

## Overview

TRIBE v2 predicts brain responses to video, audio, and text. This system adds a **persona layer** that adapts predictions based on user personality traits.

```python
# Install dependencies (uncomment if needed)
# !pip install torch numpy
# !pip install tribev2  # For full TRIBE v2 integration
```

```python
from tribev2_persona import PersonaReactionPipeline
from tribev2_persona.models import PersonaLibrary, ReactionType

# Initialize the pipeline
pipeline = PersonaReactionPipeline(device='cpu')
print("Pipeline created!")
```

## 1. Load Models

We can run in demo mode (without TRIBE v2) to test the persona system, or with full TRIBE v2 for real brain predictions.

```python
# Demo mode - doesn't require TRIBE v2
pipeline.load_models(load_tribe=False)
print("Models loaded in demo mode!")
```

## 2. Available Personas

Pre-built personas representing different user types.

```python
# List available personas
print("Available personas:")
for name in PersonaLibrary.list_personas():
    print(f"  - {name}")
```

## 3. Predict Reactions

Get reaction predictions for different personas.

```python
# Get persona
persona = PersonaLibrary.get_persona("creative")

# Predict (demo mode - simulated brain features)
result = pipeline.demo_without_tribe(persona)

# Print summary
print(pipeline.summarize_reactions(result['reactions']))
```

## 4. Generate Insights

Get human-readable insights from the predictions.

```python
# Generate insights
insights = pipeline.generate_insights(result['reactions'], result['persona'])

print("Insights for 'creative' persona:")
for key, value in insights.items():
    print(f"\n{key.upper()}:")
    print(f"  {value}")
```

## 5. Compare Personas

Compare how different personas react to the same content.

```python
# Compare personas
personas_to_compare = ["analytical", "creative", "emotional", "social"]

print("=" * 60)
print("PERSONA COMPARISON")
print("=" * 60)

for persona_name in personas_to_compare:
    persona = PersonaLibrary.get_persona(persona_name)
    result = pipeline.demo_without_tribe(persona)
    
    # Get key metrics
    attention = next(r.score for r in result['reactions'] if r.reaction_type == ReactionType.ATTENTION)
    engagement = next(r.score for r in result['reactions'] if r.reaction_type == ReactionType.ENGAGEMENT)
    emotion = next(r.score for r in result['reactions'] if r.reaction_type == ReactionType.EMOTION_AROUSAL)
    memory = next(r.score for r in result['reactions'] if r.reaction_type == ReactionType.MEMORY_ENCODING)
    
    print(f"\n{persona_name.upper()}")
    print(f"  Attention:    {'#' * int(attention * 10):10s} {attention:.2f}")
    print(f"  Engagement:   {'#' * int(engagement * 10):10s} {engagement:.2f}")
    print(f"  Emotion:      {'#' * int(emotion * 10):10s} {emotion:.2f}")
    print(f"  Memory:       {'#' * int(memory * 10):10s} {memory:.2f}")
```

## 6. Custom Persona

Create your own persona with custom traits.

```python
# Create custom persona
custom = PersonaLibrary.create_custom(
    openness=0.85,
    conscientiousness=0.6,
    extraversion=0.4,
    learning_style='visual',
    emotional_range='high'
)

# Predict
result = pipeline.demo_without_tribe(custom)

print("Custom Persona Traits:")
print(f"  Openness: {custom.openness}")
print(f"  Conscientiousness: {custom.conscientiousness}")
print(f"  Extraversion: {custom.extraversion}")
print(f"  Visual Learner: {custom.visual_learner}")
print(f"  Emotional Sensitivity: {custom.emotional_sensitivity}")

print("\n" + pipeline.summarize_reactions(result['reactions']))
```
